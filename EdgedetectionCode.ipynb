```python
def edge_detection(black_n_white, filter, activation = 'relu'):
  img_height, img_width = black_n_white.shape
  output = np.zeros((img_height,img_width))
  black_n_white = np.pad(black_n_white, 1)
  for h in range(img_height):
    vert_start = h
    vert_end = vert_start + 3
    for w in range(img_width):
      horiz_start = w
      horiz_end = horiz_start + 3
      a_slice = black_n_white[vert_start:vert_end,horiz_start:horiz_end]
      product = np.multiply(filter,a_slice)
      output[h,w] = np.sum(product)
  if activation == 'relu':
    output = np.maximum(output, 0)
  return output

x = np.ones((3,1))
y = np.zeros((3,1))
filter1 = np.block([-x,y,x])
filter2 = np.array([[0,-1,0],
                    [-1,4,-1],
                    [0,-1,0]])
filter3 = np.array([[1,0,-1],
                    [2,0,-2],
                    [1,0,-1]])
filter4 = np.array([[1,2,1],
                    [0,0,0],
                    [-1,-2,-1]])
```
